# Контест по сортировкам

## Задача I
Единственная задача, которая реально вызвала много трудностей на этом контесте. В задаче требовалось проверить массив на выполнение одного из свойств двоичной кучи. Из-за большого N возникла проблема с вылетом по памяти. Тут скорее косяк в том, что это питон, т.к. в других языках никогда не было проблем в создании массива интов такой длины. Да и по времени это укладывалось бы в секунду. Тут же пришлось искать более хитрые пути. Один из них, это создать массив вещественных чисел, которые при тех же ограничениях спокойно влазили в память(хз почему тут так). Другой путь, читать массив на ходу и, по хорошему, хранить в памяти только 2 предыдущий слой дерева и текущий. не зависимо от того, как именно считывать значения, тут просто надо пройтись по всем элементам массива и проверить, что наш элемент под номером i не больше своего предка под номером (i-1)//2. Как только мы находим элемент, превосходящий предка, то просто выдаем ответ NO и выходим из программы. Иначе просто в конце выводим YES. Оба варианта решения за O(N), но имеют разную константу.

```python
# Вариант решения с через float
n = int(input())
a = [float(input()) for _ in range(n)]
for i in range(1, n):
  if a[i] > a[(i-1)//2]:
    print("NO")
    exit(0)
print("YES")
```

```python
# "Ленивый" вариант более экономного решения.
# В случае, если наш массив - куча, то в памяти у нас кажется весь массив.
n = int(input())
a = [int(input())]
for i in range(1, n):
  a.append(int(input()))
  j = i//2 if i % 2 == 1 else i//2 - 1
  if a[i] > a[j]:
    print("NO")
    exit(0)
print("YES")
```
