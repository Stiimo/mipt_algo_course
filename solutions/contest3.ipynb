## Задача G.
В данной задаче от вас требовалось написать функцию, которая вычисляет наибольший по площади прямоугольник, который можно построить на гистограмме с заданными столбиками.<br>
При заданных ограничениях на N задача очень легко решается перебором границ прямоугольника. Мы перебираем от 0 до N-1 левую границу, от левой границы до N-1 перебираем правую. По пути вычисляем самый низкий столбик на таком отрезке. Ну и на каждом шаге сраниваем получившийся прямоугольник с наибольшим найденным на предыдущих шагах. Такое решение работает за O(N<sup>2</sup>)

```python
def largest_rect(n, a):
    ans = 0
    for i in range(n):
        h = 1000000
        for j in range(i, n):
            if a[j] < h:
                h = a[j]
            ans = ans if ans > (j-i+1)*h else (j-i+1)*h
    return ans
```

## Задача H.
В данной задаче нам надо найти число трибоначчи, превосходящее заданное. По хорошему, тут нам достаточно хранить в памяти только три последних числа, т.к. следующее число вычисляется через них. На каждом шаге просто вычисляем новое число по рекуррентной формуле из условия, выкидываем из тройки наименьшее число и получаем новую тройку чисел. Если наше новое число больше заданного, то его **номер** есть наш ответ. Такое решение работает за линейное время.

```python
a, b, c = 0, 0, 1
i = 2
x = int(input())
while True:
    if c > x:
        print(i)
        break
    a, b, c = b, c, a+b+c
    i += 1
```

## Задача I.
Условие данной задачи сводится к тому, что нам надо найти максимум в последовательности, из которой выкинули 5 последних чисел. Проблема возникает в том, что длина последовательности может достигать 10<sup>9</sup>. Из этого следует, что нам придется хранить в памяти только часть последовательности, а именно максимум и отброшенные 5 элементов. Как только нам на вход приходит ненулевое число, мы самое первое из отброшенных "прикрепляем" к последовательности, обновляя максимум по необходимости. И так, пока не встретим на входе 0. Решение работает за линию.

```python
a = [int(input()) for _ in range(6)]
while True:
    x = int(input())
    if x == 0:
        break
    a[0] = a[0] if a[1] < a[0] else a[1]
    a[1:] = a[2:]
    a.append(x)
print(a[0])
```

## Задача J.
Задача сама по себе имеет немного запутанное условие, что и было целью. На деле, тут очень простое решение. Во первых, заметим, что мы не можем уместить в памяти все дощечки забора. Но нам и не надо, нас макимум будет интересовать только 100 из них. Во вторых, не стоит нарямую красить забор, нам достаточно просто хранить эти события в виде тройки чисел (l<sub>i</sub>, r<sub>i</sub>, c<sub>i</sub>). Прочитав весь массив событий, мы начинаем обрабатывать наши M запросов. Для каждого из них считываем номер дощечки и пробегаемся по событиям в обратном порядке. Если наша дощечка поподает в событие, мы тут же прекращаем поиск и печатаем цвет. Если наша дощечка ни разе не попала в событие, то просто печатаем для нее 0.  

```python
n = int(input())
events = [[int(input()) for _ in range(3)] for _ in range(n)]
m = int(input())
for _ in range(m):
    x = int(input())
    c = 0
    for e in reversed(events):
        if e[0] <= x <= e[1]:
            c = e[2]
            break
    print(c, end=" ")
```
