## Задача A
В данной задаче от вас требовалось, по факту, реализовать ДП с ограничениями на кол-во переходов (максимальная глубина рекурсии). Тут было два варианта: либо считать от начала, останавливаясь, дойдя до целевой клетки или сделав 2 перехода, либо считать аналогичным образом с конца. Оба варианта хорошо решаются с помощью "ленивой динамики". Тут будет приведен пример решения от начала с мемоизацией (которая, по факту, тут не сильно нужна, если делать вывод прямо внутри функции и тут же завершать программу с помощью exit(0)).

```python
def count(d, x, y):
    for i, j in zip((-2, -2, -1, 1, 2, 2, 1, -1), (-1, 1, 2, 2, 1, -1, -2, -2)):
        if  0 <= x+i <= 7 and 0 <= y+j <= 7 and d[x][y] != 2 and d[x+i][y+j] == -1:
            d[x+i][y+j] = d[x][y] + 1
            count(d, x+i, y+j)


if __name__ == "__main__":
    x, y = int(input()), int(input())
    u, v = int(input()), int(input())
    d = [[-1]*8 for _ in range(8)]
    d[x-1][y-1] = 0
    count(d, x-1, y-1)
    print(d[u-1][v-1] if d[u-1][v-1] < 3 else -1)
```

## Задача E-1
Задача на подсчет кол-ва решений, однако тут есть ограничение, что мы не можем поставить подряд 3 и более контейнеров типа А. Для решения задачи определим состояния в ней как набор двух параметров: длина последовательности и кол-во опасных контейнеров, стоящих в конце последовательности. Например, последовательность ABABAA имеет праметры (6, 2). Пусть у нас есть последовательность длины i-1. Построим из нее последовательность длины i. Если мы добавляем контейнер B, то получаем состояние (i, 0), а рекуррентное соотношение для состояния: d<sub>i0</sub> = &sum; d<sub>(i-1)j</sub>, j=0..2. Если добавляем контейнер А, то получаем состояние (i, 1) или (i, 2). Для первого случая d<sub>i1</sub> = d<sub>(i-1)0</sub>, а для второго - d<sub>i2</sub> = d<sub>(i-1)1</sub>. Граничные условия у нас будут 1, 0, 0 для пустой последовательности и значению второго параметра 0, 1, 2 соответственно. Ответ будет равняться &sum; d<sub>nj</sub>, j=0..2.

```python
if __name__ == "__main__":
    n = int(input())
    d = [[0]*3 for _ in range(n+1)]
    d[0][0] = 1
    for i in range(1, n+1):
        d[i][0] = d[i-1][0]+d[i-1][1]+d[i-1][2]
        d[i][1] = d[i-1][0]
        d[i][2] = d[i-1][1]
    print(d[n][0]+d[n][1]+d[n][2])
```

## Задача I
В этой задаче довольно своеобразное решение. У нее есть несколько способов решения, но идейно их можно свести к одному. Определеим состояние в нашей задаче: у нас есть остаток i, который мы хотим разменять, используя первые j разменных монет. Тогда d<sub>ij</sub> будет означать количество способов сделать такой размен. Заметим, что d<sub>0j</sub> = 1, а для всех i > 0 d<sub>i0</sub> = 0. Рекуррентное соотношения будет следующее: d<sub>ij</sub> = d<sub>(i-c[j])j</sub> + d<sub>i(j-1)</sub>. Разберемся в соотношении. Слагаемое d<sub>(i-c[j])j</sub> означает, что мы используем монету с номиналом c[j] для размена, уменьшая остаток. При этом j остается неизменным, т.к. мы еще раз попытаемся использовать монету с таким номиналом. Стоит учесть, что это слагаемое появляется только тогда, когда монета не превышает остаток. Слагаемое d<sub>i(j-1)</sub> означает, что мы больше не пытаемся использовать монету c[j], и поэтому уменьшаем список наших разменных монет на 1. Ответ будет лежать в d<sub>nm</sub>, где n - начальная сумма, m - длина изначального списка монет.

```python
def make_exchange(money, coins):
    d = [[0]*(len(coins)+1) for _ in range(money+1)]
    d[0] = [1]*(len(coins)+1)
    for i in range(1, money+1):
        for j in range(1, len(coins)+1):
            d[i][j] = d[i][j-1]
            if i - coins[j-1] >= 0:
                d[i][j] += d[i-coins[j-1]][j]
    return d[money][len(coins)]
```
