# Система контроля версий git

Это система, регистрирующая изменения в файлах и позволяющая вернуться к их старым версиям. Тут я не буду углубляться во все тонкости этой системы. Во время работы над проектом вы будете использовать git. Ваш репозиторий будет находиться на https://github.com.

## Начало работы с git
Во первых необходимо установить git к себе на комп. Для пользователей Windows в интернете должен быть установщик, пользователи Unix систем могут воспользоваться менеджером пакетов. После установки вы сможете пользоваться командами git из консоли. К тому же PyCharm имеет встроенное меню для пользователей git.

После установки вам надо будет выгрузить созданный вами репозиторий. Для этого нужна команда **git clone**:
```
git clone https://github.com/Stiimo/mipt_python3_course.git mipt_python3_course
```
Последним аргументом указывается название корневой папки вашей локальной копии, ее имя не обязательно совпадает с названием репозитория. Папка создается в директории, из которой была запущена команда. После того, как вы выгрузили репозиторий, все вносимые локально изменения будут отслеживаться.

## Сохранение изменений на сервер
После того, как вы внесли достаточное на ваш взгляд количество измнений в проект, имеет смысл сохранить его текущую версию. В случае неудачных последующих изменений можно будет легко вернуться к нужной вам версии. Для этого вам необходимо применить изменения (сделать commit-сообщение) и отправить на сервер. Сначала надо выбрать, какие файлы добавить в commit. Для этого можно воспользоваться командой **git add <файлы>**. Тут может быть указан как один файл, так и несколько через пробел. В названиях можно использовать спец символы * и ?, вместо которых система сможет подставлять разные символы. Например, чтобы добавить в commit все измененные файлы:
```
git add *
```
После этого можно выполнить команду **git commit**. При запуске команды вам будет предложено ввести поясняющее сообщение, пишите их осмысленно, чтобы потом не вспоминать с трудом, что тут имелось в виду. Выполнив эту команду, локально появится очередная "точка сохранения", к которой можно будет возвращаться. Она будет включать те измнения, каторые были добавлены командой **git add**. Чтобы изменения появились на сервере, выполните команду **git push**.
